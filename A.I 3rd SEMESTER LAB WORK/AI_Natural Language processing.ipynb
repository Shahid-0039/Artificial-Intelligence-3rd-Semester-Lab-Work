```python
my_String="Let's write RegEx!  Won't that be fun?  I sure think so.  Can you find 4 sentences?  Or perhaps, all 19 words?"
print(my_String)
```

```python
import re
```

```python
# Write a pattern to match sentence endings: sentence_endings
sentence_endings = r"[.?!]"

# Split my_string on sentence endings and print the result
print(re.split(sentence_endings, my_String))
```

```python
# Find all letters that start with a capital letter in my_string 
capitalized_words = r"[A-Z]"
print(re.findall(capitalized_words, my_String))
```

```python
# Find all letters that start with a capital letter in my_string 
small_words = r"[a-z]"
print(re.findall(small_words, my_String))
```

```python
# Find all letters that start with a capital letter in my_string 
capitalized_words = r"[A-Z][a-z][a-z]'\w+"
print(re.findall(capitalized_words, my_String))
```

```python
# Find all words that start with a capital letter in my_string 
capitalized_words = r"[A-Z]\w+"
print(re.findall(capitalized_words, my_String))
```

```python
# Split my_string on spaces and print the result
spaces = r"\s+"
print(re.split(spaces, my_String))
```

```python
# Split my_string on ' and print the result
spaces = r"'"
print(re.split(spaces, my_String))
```

```python
# Find all digits in my_string and print the result
digits = r"\d+"
print(re.findall(digits, my_String))
```

```python
new_String="123Someone23 and How are you? 54Hello5"
```

```python
#Start with one or more digits, then one or more words in between and end with one or more digits
pattern=r"\d+[a-zA-Z]+\d+" #or  r"\d+w+\d+"
re.findall(pattern,new_String)
```

```python
#Start with 5, we use b to limit the boundary
pattern=r"\b5\w+"
re.findall(pattern,new_String)
```

```python
#Ends with 5
pattern=r"\d+\w+5\b"
re.findall(pattern,new_String)
```

```python
#Write down regular expression for matching CNIC
cnic="45204-6147185-9, 23233-44-12, +923068372744"
CNIC_PATTERN=r"\d{5}-\d{7}-\d"
print(re.findall(CNIC_PATTERN,cnic))
```

```python
#Write down regular expression for matching Cell Number"
cnic="45204-6147185-9, 23233-44-12, +923068372744"
CNIC_PATTERN=r"\+92\d{10}"
print(re.findall(CNIC_PATTERN,cnic))
```

<H2>Advanced Tokenization with NLTK</H2>

```python
#Tokenizer for tweets with hashtags
# Import the necessary modules
from nltk.tokenize import regexp_tokenize
from nltk.tokenize import TweetTokenizer
```

```python
tweets=['This is the best #nlp exercise ive found online! #python',
 '#NLP is super fun! <3 #learning',
 'Thanks @datacamp :) #nlp #python']
```

```python
# Define a regex pattern to find hashtags: pattern1
pattern1 = r"\#\w+"
# Use the pattern on the first tweet in the tweets list
hashtags = regexp_tokenize(tweets[0], pattern1)
print(hashtags)
```

```python
# Write a pattern that matches both mentions (@) and hashtags
pattern2 = r"(\@[a-zA-Z]+|\#\w+)"
# Use the pattern on the last tweet in the tweets list
mentions_hashtags = regexp_tokenize(tweets[-1], pattern2)
print(mentions_hashtags)
```

```python
tknzr=TweetTokenizer()
all_tokens=[]
for t in tweets:
    all_tokens.append(tknzr.tokenize(t))
print(all_tokens)
```

```python
from collections import Counter
# Create the bag-of-words: bow
bow = Counter(lower_tokens)

# Print the 10 most common tokens
print(bow.most_common(10))
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
# Retain alphabetic words: alpha_only
alpha_only=[]
for t in lower_tokens:
    if t.isalpha():
        alpha_only.append(t)
```

```python
# Remove all stop words: no_stops
no_stops=[]
for t in alpha_only:
    if t not in stopwords.words('english'):
        no_stops.append(t)
```

```python
# Instantiate the WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_tokens=[]
for t in no_stops:
    lemmatized_token=wordnet_lemmatizer.lemmatize(t)
    lemmatized_tokens.append(lemmatized_token)
```

```python
# Create the bag-of-words: bow
bow = Counter(lemmatized_tokens)

# Print the 10 most common tokens
print(bow.most_common(10))
```
